**Observações**:

A tabela resulta de dois labels (na verdade visualizações): "PhotoObj" que contém dados fotométricos e "SpecObj" que contém dados espectrais. Iniciaremos detalhando "PhotoObj":

- objid = identificador de objeto
- ra = Ascensão Reta J2000 (banda r)
- dec = Declinação J2000 (banda r)
- O sistema de magnitude astronômica Thuan-Gun: u, g, r, i, z representam a resposta das 5 bandas do telescópio.
- run = número de execução
- rerun = número de reexecução
- camcol =  camêra coluna
- fiel = número do campo

Estes últimos 4 são recursos que descrevem um campo dentro de uma imagem obtida pelo SDSS. Um campo é basicamente uma parte da imagem inteira correspondendo a 2048 por 1489 pixels. Ele pode ser representado pelo número run, que identifica a verificação específica; a camcol, que é basicamente um range entre 1 a 6, números que identificam a linha de varredeura dentro da execução; o número do campo, que normalmente começa em 11 (após um tempo alavancagem inicial) e pode chegar a 800 para execuções particularmente longas; e por fim, o rerun que especifica como a imagem foi processada.
Então, temos agora dados "SpecObj":

- specobjid = identificador do objeto
- classe = classe de objeto (objeto de galáxia, estrela ou quasar)
- redshift = Redshift final
- placa = número da placa
- mjd = MJD de observação
- fibreid = fibra ID

O desvio para o vermelho (redshift) acontece quando a luz ou outra radiação eletromagnética de um objeto é aumentada em comprimento de onda ou deslocada para a extremidade vermelha do espectro. Observe que nas 5 primeiras colunas mostradas, existem redshifts negativos, o que significa na verdade: blueshift, ou seja, desvio para o azul. Em outras palavras, o redshift para esse dataset indica que os objetos estão se afastando do observador e no blueshift, estão se aproximando.
Cada exposição espectroscópica emprega uma grande e fina placa de metal circular que posiciona as fibras ópticas por meio de orifícios perfurados nas localizações das imagens no plano focal do telescópio. Essas fibras então alimentam os espectrógrafos. Cada placa possui um número de série exclusivo, que é chamado de placa em visualizações como SpecObj no CAS.

Para se estudar a cor de uma estrela, mede-se sua magnitude através
de diferentes filtros (bandas). Um telescópio equipado com esses filtros pode medir as magnitudes
aparentes (mU, mB , mV) nas bandas do ultravioleta, azul e visível. Para simplificar, denota-se mU como U, simplesmente. Os sistemas fotométricos também se estendem para outras faixas espectrais
como o vermelho (ex: R, I).

**O índice de cor é a diferença entre magnitudes aparentes (brilhos) de
duas bandas.**

Por convenção, fazemos:
(banda mais azul – banda mais vermelho).

(B-V) < 0
Estrela quente, azulada,
tem índice de cor
negativo

(B-V) > 0
Estrela fria, avermelhada,
tem índice de cor positivo

Os índices de cor têm grande importância em astrofísica, pois são uma
forma de se medir a temperatura da fotosfera com relativa facilidade.

                        Testamos a COMBINAÇÃO DE ATRIBUTOS

Fiberid, objid, run, rerun, camcol, field, plate serão descartados porque estão mais relacionados a como os dados são organizados e provavelmente não afetam a classificação do objeto. Mjd é mantido porque está relacionado às estações do ano e afeta quais objetos são visíveis no céu noturno. Durante o outono e a primavera, objetos extra-galácticos como galáxias e QSO são mais propensos a serem fotografados e durante o verão e inverno objetos intra-galácticos são mais visíveis, como estrelas. Specobjid será descartado porque há mais interesse em classificar com base em quantidades mensuráveis, como a mangitude e a localização do objeto.

A Ascensão Reta (ra) será descartada porque assumimos um universo homogêneo. A declinação (dec) será mantida porque há uma maior probabilidade de encontrar galáxias e QSO fora do plano galáctico (DEC ~ -29), portanto, a declinação do objeto deve ser considerada.
